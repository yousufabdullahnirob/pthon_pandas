```python
import pandas as pd


df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 88, 92, 95]
})


var_new = df.groupby('Name')


for x, y in var_new:
    print(x)
    print(y)
```
