```python
import pandas as pd
```

```python
var = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'var': [10, 20, 30, 40, 50],
    'other_col': [100, 200, 300, 400, 500]
})
var
```

```python
# pivot() does not support margins, using pivot_table() instead
var.pivot_table(index="id", columns="var", values="other_col", margins=True, aggfunc="sum")
```
