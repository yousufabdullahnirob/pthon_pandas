```python
import pandas as pd

sr1 = pd.DataFrame({'A':[1,2,3,4], 'B':[5,6,7,8]})
sr2 = pd.DataFrame({'C':[5,6,7,8], 'D':[5,6,7,8]})
pd.concat([sr1, sr2], axis=1,join="inner")
```
